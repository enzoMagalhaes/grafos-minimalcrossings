```python
import random

# Generate edges programmatically with randomness
num_nodes = 500  # Total number of nodes
edges = []

# Add random connections, ensuring no self-loops
for i in range(1, num_nodes + 1):
    num_edges = 2  # Define the number of random connections per node
    connected_nodes = set()
    while len(connected_nodes) < num_edges:
        target = random.randint(1, num_nodes)
        if target != i and target not in connected_nodes:  # Ensure no self-loops or duplicate edges
            connected_nodes.add(target)
            edges.append((i, target))


run_and_plot_graph_analysis(edges)
```
